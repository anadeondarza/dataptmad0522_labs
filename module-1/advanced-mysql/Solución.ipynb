```python
query1_step1 = '''
SELECT
t.title_id AS TITLE_ID,
ta.au_id AS AUTHOR_ID,
(t.advance*(CAST(ta.royaltyper AS FLOAT)/100)) AS ADVANCE,
(t.price * s.qty * (CAST(t.royalty AS FLOAT) / 100) * (CAST(ta.royaltyper AS FLOAT) / 100))  as SALES_ROYALTY

from titles as t
INNER JOIN titleauthor as ta ON t.title_id = ta.title_id
INNER JOIN sales as s ON t.title_id = s.title_id
'''
```

```python
query1_step1 = pd.read_sql_query(query1_step1, engine)
query1_step1
```
